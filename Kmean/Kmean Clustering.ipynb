```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Kmean2.csv')
df.head()
```

```python
X = df['x']
Y = df['y']
plt.scatter(X,Y)
```

```python
kmeans = KMeans(n_clusters=4).fit(df)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python
df['labels'] = labels

group1 = df[df['labels']==1].plot( kind='scatter', x='x', y='y', color='Green' )
group2 = df[df['labels']==0].plot( kind='scatter', x='x', y='y', color='Gray', ax=group1 )
group2 = df[df['labels']==2].plot( kind='scatter', x='x', y='y', color='Blue', ax=group1  )
group2 = df[df['labels']==3].plot( kind='scatter', x='x', y='y', color='Blue', ax=group1 )
group2 = df[df['labels']==4].plot( kind='scatter', x='x', y='y', color='Blue', ax=group1 )

plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=150, linewidths = 5, color='Red')

plt.show()
```

```python

cols = ['r', 'g', 'b', 'y', 'gray', 'black']
group1 = df[df['labels']==0].plot( kind='scatter', x='x', y='y', color=cols[0] )
for i in range (0,len(centroids)):
    df[df['labels']==i].plot( kind='scatter', x='x', y='y', color=cols[i], ax=group1 )
plt.scatter(centroids[:, 0],centroids[:, 1], marker = "x", s=70, linewidths = 5, color='black')
```

```python
for i in range (0,len(centroids)):
    df1 = df[df['labels']==i]

    xd = df1['x']
    yd = df1['y']

    plt.scatter(xd,yd, color =cols[i])
    plt.plot(xd,yd, color =cols[i])
```
