## Linear Regression

to develop a programme that reads the area and the price of lands and train a model to predict the price of unknown land based on its area

Regression
Linear Regression
Linear Regression is a supervised machine learning algorithm.
It tries to find out the best linear relationship that describes the data you have.
It assumes that there exists a linear relationship between a dependent variable and independent variable(s).
The value of the dependent variable of a linear regression model is a continuous value i.e. real numbers


```python
import pandas as pd # Dealing with CSV 
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/kaggle/input/az-land-price/landPrice.csv') # CSV means Comma Separated Value
df.describe()
```

```python
df
```

```python
X = df['Area']
Y = df['Price']
Y
```

```python
plt.scatter(X,Y)
```

```python
from scipy import stats
a, b, x1, x2, x3 = stats.linregress(X,Y)

def f(X):
  return a*X + b

mymodel = list (map(f,X))
plt.scatter(X,Y) # the original data
plt.plot (X, mymodel) # the line
plt.show()
print (a,b)
```

```python
pr = f(120)

print (pr)
```

# Polynomial Regression

```python

X = df['Area']
Y = df['Price']
X = np.array(X)
Y = np.array(Y)
myline = np.linspace(100, 600, 10)
plt.scatter(X, Y)


mymodel = np.poly1d(np.polyfit(X, Y, 4))
plt.scatter(X, Y)
plt.plot(myline, mymodel(myline))
plt.show()

price = mymodel(120)
print (price)
```

# Overfitting

```python
X = df['Area']
Y = df['Price']
X = np.array(X)
Y = np.array(Y)
myline = np.linspace(100, 600, 10)
plt.scatter(X, Y)


mymodel = np.poly1d(np.polyfit(X, Y, 12))
plt.scatter(X, Y)
plt.plot(myline, mymodel(myline))
plt.show()

price = mymodel(120)
print (price)
```
